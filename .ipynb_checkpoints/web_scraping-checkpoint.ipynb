```python
#Import these libraries
import requests
from bs4 import BeautifulSoup
```

```python
#URL Extraction
url = "https://boston.craigslist.org/search/sof"
```

```python
response = requests.get(url)
```

```python
response #200 means Great, your internet connection works
```

```python
#Now we need to extract the source code of the web page
data = response.text
```

```python
data
```

```python
#Pass the code to BS4 
soup= BeautifulSoup(data, "html.parser")
```

```python
tags = soup.find_all('a')
```

```python
tags #Test with web page
```

```python
for tag in tags:
    print(tag.get('href'))
```

```python
titles = soup.find_all("a", {"class": "result-title"})
```

```python
for title in titles:
    print(title.text)
```

```python
adresses  = soup.find_all("span", {"class": "result-hood"})
```

```python
for address in adresses:
    print(address.text)
```

```python
jobs = soup.find_all('p', {'class': 'result-info'})
```

```python
for job in jobs:
    title = job.find('a', {'class': 'result-title'}).text
    location_tag = job.find('span', {'class': 'result-hood'})
    location = location_tag.text[2:-1] if location_tag else "N/A"
    date = job.find('time', {'class': 'result-date'}).text
    link = job.find('a', {'class': 'result-title'}).get('href')
    print('Job Title:', title, '\nLocation:', location, '\Date:', date, '\Link: ', link, '\n---')
```
